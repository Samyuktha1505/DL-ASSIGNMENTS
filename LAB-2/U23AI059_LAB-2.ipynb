```python
# ================================
# DL Practical-2 (MNIST)
# CNN + MLP + Experiments + Graphs + Comparison Table
# ================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print("TensorFlow Version:", tf.__version__)

# -------------------------------
# Load MNIST Dataset
# -------------------------------
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

# Reshape for CNN (28,28,1)
x_train_cnn = np.expand_dims(x_train, -1)
x_test_cnn  = np.expand_dims(x_test, -1)

# One-hot encoding (for some models if needed)
num_classes = 10

# -------------------------------
# Helper: Build CNN Model
# -------------------------------
def build_cnn_model(activation="relu",
                    optimizer_name="adam",
                    use_bn=False,
                    dropout_rate=0.25):
    
    model = keras.Sequential()
    model.add(layers.Input(shape=(28,28,1)))

    # Conv Layer 1
    model.add(layers.Conv2D(32, (3,3), padding="same"))
    if use_bn:
        model.add(layers.BatchNormalization())
    model.add(layers.Activation(activation))

    # Conv Layer 2
    model.add(layers.Conv2D(64, (3,3), padding="same"))
    if use_bn:
        model.add(layers.BatchNormalization())
    model.add(layers.Activation(activation))

    # Pooling
    model.add(layers.MaxPooling2D(pool_size=(2,2)))

    # Dropout
    model.add(layers.Dropout(dropout_rate))

    # Flatten + Dense
    model.add(layers.Flatten())
    model.add(layers.Dense(128))
    if use_bn:
        model.add(layers.BatchNormalization())
    model.add(layers.Activation(activation))

    # Output
    model.add(layers.Dense(10, activation="softmax"))

    # Optimizer selection
    if optimizer_name.lower() == "sgd":
        opt = keras.optimizers.SGD(learning_rate=0.01)
    elif optimizer_name.lower() == "momentum":
        opt = keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
    elif optimizer_name.lower() == "adam":
        opt = keras.optimizers.Adam(learning_rate=0.001)
    else:
        raise ValueError("Optimizer must be: sgd / momentum / adam")

    model.compile(optimizer=opt,
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model

# -------------------------------
# Helper: Build MLP Model
# -------------------------------
def build_mlp_model(hidden_layers=[256,128],
                    activation="relu",
                    optimizer_name="adam",
                    use_bn=True,
                    dropout_rate=0.0):
    
    model = keras.Sequential()
    model.add(layers.Input(shape=(28,28)))
    model.add(layers.Flatten())  # 784

    for units in hidden_layers:
        model.add(layers.Dense(units))
        if use_bn:
            model.add(layers.BatchNormalization())
        model.add(layers.Activation(activation))
        if dropout_rate > 0:
            model.add(layers.Dropout(dropout_rate))

    model.add(layers.Dense(10, activation="softmax"))

    # Optimizer selection
    if optimizer_name.lower() == "sgd":
        opt = keras.optimizers.SGD(learning_rate=0.01)
    elif optimizer_name.lower() == "momentum":
        opt = keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
    elif optimizer_name.lower() == "adam":
        opt = keras.optimizers.Adam(learning_rate=0.001)
    else:
        raise ValueError("Optimizer must be: sgd / momentum / adam")

    model.compile(optimizer=opt,
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model

# -------------------------------
# Helper: Train + Store History
# -------------------------------
def train_and_evaluate(model, xtr, ytr, xte, yte, epochs=10, batch_size=128, name="Experiment"):
    print(f"\n================ {name} ================")
    history = model.fit(
        xtr, ytr,
        validation_data=(xte, yte),
        epochs=epochs,
        batch_size=batch_size,
        verbose=1
    )
    test_loss, test_acc = model.evaluate(xte, yte, verbose=0)
    print(f"✅ Final Test Accuracy: {test_acc:.4f}")
    return history, test_acc

# -------------------------------
# Plot Training Curves
# -------------------------------
def plot_history(histories_dict, metric="accuracy", title=""):
    plt.figure(figsize=(10,6))
    for label, hist in histories_dict.items():
        plt.plot(hist.history[metric], label=f"{label} Train")
        plt.plot(hist.history["val_"+metric], linestyle="--", label=f"{label} Val")
    plt.xlabel("Epochs")
    plt.ylabel(metric.capitalize())
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_loss(histories_dict, title=""):
    plt.figure(figsize=(10,6))
    for label, hist in histories_dict.items():
        plt.plot(hist.history["loss"], label=f"{label} Train Loss")
        plt.plot(hist.history["val_loss"], linestyle="--", label=f"{label} Val Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# ==========================================================
# TASK 1: Activation Function Challenge (CNN)
# Compare Sigmoid vs Tanh vs ReLU (Optimizer fixed = SGD)
# ==========================================================
activation_experiments = [
    {"activation":"sigmoid", "optimizer":"sgd", "epochs":10},
    {"activation":"tanh",    "optimizer":"sgd", "epochs":10},
    {"activation":"relu",    "optimizer":"sgd", "epochs":10},
]

task1_histories = {}
results_table = []

for exp in activation_experiments:
    model = build_cnn_model(activation=exp["activation"],
                            optimizer_name=exp["optimizer"],
                            use_bn=False,
                            dropout_rate=0.25)
    h, acc = train_and_evaluate(model, x_train_cnn, y_train, x_test_cnn, y_test,
                                epochs=exp["epochs"],
                                name=f"TASK1 CNN ({exp['activation']} + {exp['optimizer']})")
    task1_histories[f"{exp['activation']}+{exp['optimizer']}"] = h
    results_table.append(["TASK1-CNN", exp["activation"], exp["optimizer"], exp["epochs"], acc])

plot_history(task1_histories, metric="accuracy", title="TASK 1: CNN Accuracy Curves (Sigmoid vs Tanh vs ReLU)")
plot_loss(task1_histories, title="TASK 1: CNN Loss Curves (Sigmoid vs Tanh vs ReLU)")

# ==========================================================
# TASK 2: Optimizer Showdown (Activation fixed = ReLU)
# Compare SGD vs Momentum vs Adam (CNN)
# ==========================================================
optimizer_experiments = [
    {"activation":"relu", "optimizer":"sgd",      "epochs":10},
    {"activation":"relu", "optimizer":"momentum", "epochs":10},
    {"activation":"relu", "optimizer":"adam",     "epochs":10},
]

task2_histories = {}

for exp in optimizer_experiments:
    model = build_cnn_model(activation=exp["activation"],
                            optimizer_name=exp["optimizer"],
                            use_bn=False,
                            dropout_rate=0.25)
    h, acc = train_and_evaluate(model, x_train_cnn, y_train, x_test_cnn, y_test,
                                epochs=exp["epochs"],
                                name=f"TASK2 CNN ({exp['activation']} + {exp['optimizer']})")
    task2_histories[f"{exp['activation']}+{exp['optimizer']}"] = h
    results_table.append(["TASK2-CNN", exp["activation"], exp["optimizer"], exp["epochs"], acc])

plot_history(task2_histories, metric="accuracy", title="TASK 2: CNN Accuracy Curves (SGD vs Momentum vs Adam)")
plot_loss(task2_histories, title="TASK 2: CNN Loss Curves (SGD vs Momentum vs Adam)")

# ==========================================================
# TASK 3: BN + Dropout Scenarios (CNN)
# Scenarios:
# 1) WITHOUT BN and WITHOUT Dropout
# 2) Without BN, dropout=0.1
# 3) With BN, dropout=0.25
# ==========================================================
task3_experiments = [
    {"use_bn":False, "dropout":0.0,  "activation":"relu", "optimizer":"adam", "epochs":10, "label":"NoBN_NoDrop"},
    {"use_bn":False, "dropout":0.1,  "activation":"relu", "optimizer":"adam", "epochs":10, "label":"NoBN_Drop0.1"},
    {"use_bn":True,  "dropout":0.25, "activation":"relu", "optimizer":"adam", "epochs":10, "label":"BN_Drop0.25"},
]

task3_histories = {}

for exp in task3_experiments:
    model = build_cnn_model(activation=exp["activation"],
                            optimizer_name=exp["optimizer"],
                            use_bn=exp["use_bn"],
                            dropout_rate=exp["dropout"])
    h, acc = train_and_evaluate(model, x_train_cnn, y_train, x_test_cnn, y_test,
                                epochs=exp["epochs"],
                                name=f"TASK3 CNN ({exp['label']})")
    task3_histories[exp["label"]] = h
    results_table.append(["TASK3-CNN", exp["activation"], exp["optimizer"], exp["epochs"], acc])

plot_history(task3_histories, metric="accuracy", title="TASK 3: CNN Accuracy Curves (BN + Dropout Scenarios)")
plot_loss(task3_histories, title="TASK 3: CNN Loss Curves (BN + Dropout Scenarios)")

# ==========================================================
# BASE TABLE REQUIREMENT FROM PDF
# CNN-1 128 Adam 10
# MLP-1 512-256-128 SGD 20
# MLP-2 256 Adam 15
# ==========================================================
print("\n\n================ PDF REQUIRED MODELS ================")

# CNN-1
cnn1 = build_cnn_model(activation="relu", optimizer_name="adam", use_bn=False, dropout_rate=0.25)
h_cnn1, acc_cnn1 = train_and_evaluate(cnn1, x_train_cnn, y_train, x_test_cnn, y_test,
                                      epochs=10, name="CNN-1 (128 FC, Adam, 10 epochs)")
results_table.append(["CNN-1", "relu", "adam", 10, acc_cnn1])

# MLP-1 (512-256-128, SGD, 20)
mlp1 = build_mlp_model(hidden_layers=[512,256,128], activation="relu", optimizer_name="sgd", use_bn=True, dropout_rate=0.0)
h_mlp1, acc_mlp1 = train_and_evaluate(mlp1, x_train, y_train, x_test, y_test,
                                      epochs=20, name="MLP-1 (512-256-128, SGD, 20 epochs)")
results_table.append(["MLP-1", "relu", "sgd", 20, acc_mlp1])

# MLP-2 (256, Adam, 15)
mlp2 = build_mlp_model(hidden_layers=[256], activation="relu", optimizer_name="adam", use_bn=True, dropout_rate=0.0)
h_mlp2, acc_mlp2 = train_and_evaluate(mlp2, x_train, y_train, x_test, y_test,
                                      epochs=15, name="MLP-2 (256, Adam, 15 epochs)")
results_table.append(["MLP-2", "relu", "adam", 15, acc_mlp2])

# ==========================================================
# Final Comparison Table Output
# ==========================================================
df = pd.DataFrame(results_table, columns=["Experiment", "Activation", "Optimizer", "Epochs", "Final Test Accuracy"])
df = df.sort_values(by="Final Test Accuracy", ascending=False).reset_index(drop=True)

print("\n\n✅ FINAL COMPARISON TABLE (Sorted by Accuracy):")
display(df)

# Plot MLP / CNN required model accuracy comparison (bar-like simple plot)
plt.figure(figsize=(8,5))
plt.plot(["CNN-1","MLP-1","MLP-2"], [acc_cnn1, acc_mlp1, acc_mlp2], marker="o")
plt.title("PDF Models: Final Test Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Test Accuracy")
plt.grid(True)
plt.show()
```
